```python
import pandas as pd
import numpy as np
```

```python
#-----------------------------Exercise 1--------------------------------#
imdb_01 = pd.read_csv('imdb-1.csv', header=None)
com402_01 = pd.read_csv('com402-1.csv', header=None)
```

```python
#Test
imdb_01[imdb_01[0] == 'donald.trump@whitehouse.gov']
```

```python
#Get rid of duplicate dates
imdb_01_dates = imdb_01.drop_duplicates(subset = [2,3])
com402_01_dates = com402_01.drop_duplicates(subset =[2,3])
```

```python
#Merge by dates
merged_dates = pd.merge(imdb_01_dates, com402_01_dates, on=2)
display(merged_dates)
```

```python
#Find hashes of movies and users
movie_hashes = merged_dates[['1_x', '1_y']].drop_duplicates(subset = '1_x')
user_hashes = merged_dates[['0_x', '0_y']].drop_duplicates(subset = '0_x')
display(movie_hashes)
```

```python
#find the hash of Trump
trump_hash = user_hashes[user_hashes['0_x']=='donald.trump@whitehouse.gov']['0_y'].values[0]
trump_table = com402_01[com402_01[0] == trump_hash]
display(trump_table)
```

```python
#Hash values of Trump movies
trump_movies_hashed = trump_table[1]
trump_movies_hashed.values
```

```python
#Get the list of movies sorted
movie_hashes[movie_hashes['1_y'].isin(trump_movies_hashed.values)].sort_values('1_x')
```

```python
#------------------------Exercise 2--------------------------#
imdb_02 = pd.read_csv('imdb-2.csv', header=None)
com402_02 = pd.read_csv('com402-2.csv', header=None)
```

```python
#Group by movie name, count number of ratings
imdb_02_counts = imdb_02.groupby(1)[1].count().sort_values(ascending = False)
com402_02_counts = com402_02.groupby(1)[1].count().sort_values(ascending = False)
frame = { 'Movie': imdb_02_counts.keys(), 'Hash': com402_02_counts.keys() } 
merged_counts = pd.DataFrame(frame)
display(merged_counts)
```

```python
#find out trump movies in the smaller IMDB database
imdb_02_group_users = imdb_02.groupby(0)[1].apply(list).reset_index(name='Movies')
trump_movies_02_imdb = imdb_02_group_users[imdb_02_group_users[0] == 'donald.trump@whitehouse.gov']
display(trump_movies_02_imdb['Movies'].values)
```

```python
#Merge hashes of the movies with the name of the movies
com_402_with_names = pd.merge(com402_02, merged_counts, left_on = 1, right_on='Hash', how='inner')

#Group by the hashed users
com402_02_group_users = com_402_with_names.groupby(0)['Movie'].apply(list).reset_index(name='Movies')
display(com402_02_group_users)
```

```python
#Return the set of movies from Com402 database which is a superset of the movies rated by Trump
trump_movies_02 = []

for movie in com402_02_group_users['Movies']:
    if(set(trump_movies_02_imdb['Movies'].values[0]).issubset(set(movie))):
        trump_movies_02 = movie
    
display(set(trump_movies_02))
```
